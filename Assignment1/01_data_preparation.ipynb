```python
from datasets import load_dataset
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.model_selection import train_test_split
```

```python
from datasets import load_dataset
ds = load_dataset("blanchon/UC_Merced")
# The dataset has a 'train' split only - extract all data
all_data = ds['train']
labels = all_data['label']
label_names = all_data.features['label'].names  # 21 class names
```

```python
from sklearn.model_selection import train_test_split
indices = list(range(len(all_data)))
train_idx, temp_idx = train_test_split(indices, test_size=0.30, stratify=labels, random_state=42)
val_labels = [labels[i] for i in temp_idx]
val_idx, test_idx = train_test_split(temp_idx, test_size=0.50, stratify=val_labels, random_state=42)
print(f"Train: {len(train_idx)}, Val: {len(val_idx)}, Test: {len(test_idx)}")
```

```python
import json
splits = {'train': train_idx, 'val': val_idx, 'test': test_idx}
with open('splits.json', 'w') as f:
    json.dump(splits, f)
print("Splits saved to splits.json")
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Find first image per class
class_samples = {}
for i in train_idx:
    lbl = all_data[i]['label']
    if lbl not in class_samples:
        class_samples[lbl] = all_data[i]['image']
    if len(class_samples) == 21:
        break
fig, axes = plt.subplots(4, 6, figsize=(18, 12))
for idx, (lbl, img) in enumerate(sorted(class_samples.items())):
    ax = axes[idx // 6][idx % 6]
    ax.imshow(img)
    ax.set_title(label_names[lbl], fontsize=8)
    ax.axis('off')
# hide unused
for idx in range(21, 24):
    axes[idx // 6][idx % 6].axis('off')
plt.suptitle('One Sample Per Class – UC Merced Land Use', fontsize=14)
plt.tight_layout()
plt.savefig('class_samples.png', dpi=100)
plt.show()
```

```python
from collections import Counter
train_counts = Counter(labels[i] for i in train_idx)
val_counts   = Counter(labels[i] for i in val_idx)
test_counts  = Counter(labels[i] for i in test_idx)
print(f"{'Class':<20} {'Train':>8} {'Val':>8} {'Test':>8} {'Total':>8}")
print('-' * 52)
for c in range(21):
    print(f"{label_names[c]:<20} {train_counts[c]:>8} {val_counts[c]:>8} {test_counts[c]:>8} {train_counts[c]+val_counts[c]+test_counts[c]:>8}")
```
